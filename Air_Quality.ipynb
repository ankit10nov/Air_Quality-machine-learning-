```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("city_day.csv")
df.info()
```

```python
df.sample(10)
```

```python
#df = df.drop(columns=['Xylene','PM10','NH3'])
df.isna().sum()
df['PM2.5']= df['PM2.5'].fillna(df['PM2.5'].mean())
df['NO']= df['NO'].fillna(df['NO'].mean())
df['NO2']= df['NO2'].fillna(df['NO2'].mean())
df['NOx']= df['NOx'].fillna(df['NOx'].mean())
df['NOx']= df['NOx'].fillna(df['NOx'].mean())

df.isna().sum()
```

```python
df['CO']= df['CO'].fillna(df['CO'].mean())
df['SO2']= df['SO2'].fillna(df['SO2'].mean())
df['O3']= df['O3'].fillna(df['O3'].mean())
df['Benzene']= df['Benzene'].fillna(df['Benzene'].mean())
df['Toluene']= df['Toluene'].fillna(df['Toluene'].mean())
df['AQI_Bucket']= df['AQI_Bucket'].fillna(df['AQI_Bucket'].mode().iloc[0])
df = df.drop('PM10',axis=1)
df = df.drop('NH3',axis=1)
df = df.drop('Xylene',axis=1)
df = df.drop('Date',axis=1)
df = df.drop('AQI',axis=1)
```

```python
df.isna().sum()
```

```python
df['AQI_Bucket'].mode().iloc[0]
```

```python

X = df.iloc[:, :-1]
print(X)

y = df.iloc[:, -1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
X_encoded = pd.get_dummies(X, columns=['City'])
X_encoded
```

```python
X_train_encoded, X_test_encoded, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train_encoded, y_train)
```

```python
yhat=model.predict(X_test_encoded)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,yhat)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
model3= RandomForestClassifier()
model3.fit(X_train_encoded, y_train)
yhat=model3.predict(X_test_encoded)
accuracy_score(y_test,yhat)
print(classification_report(y_test,yhat))
```

```python
model4=GaussianNB()
model4.fit(X_train_encoded, y_train)
yhat=model4.predict(X_test_encoded)
accuracy_score(y_test,yhat)
print(classification_report(y_test,yhat))
```

```python
model5=SVC()
model5.fit(X_train_encoded, y_train)
yhat=model5.predict(X_test_encoded)
accuracy_score(y_test,yhat)
print(classification_report(y_test,yhat))
```

```python
model6=KNeighborsClassifier()
model6.fit(X_train_encoded,y_train)
yhat=model6.predict(X_test_encoded)
accuracy_score(y_test,yhat)
print(classification_report(y_test,yhat))
```
